```python
# SELECTING 10 BEST FEATURES BASED ON LASSO

from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)

lasso.fit(x, y)

coef_magnitude = np.abs(lasso.coef_)
top_indices = np.argsort(coef_magnitude)[::-1][:10]
top_features = x.columns[top_indices]

print('Top features:', top_features)
```

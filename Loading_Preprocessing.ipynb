```python
# CHECKING NUMBER OF OUTLIERS IN THE TARGET VARIABLE

median = np.median(data['yield'])
q1, q3 = np.percentile(data['yield'], [25, 75])
iqr = q3 - q1

upper_bound = median + 1.5*iqr
lower_bound = median - 1.5*iqr

num_outliers = np.sum((data['yield'] < lower_bound) | (data['yield'] > upper_bound))
num_outliers
```

```python
# BECAUSE OF LARGE NUMBER OF OUTLIERS, REMOVING OUTLIERS IS A BAD IDEA AS IT MAY CONTAIN IMPORTANT INFORMATION THUS ITS BETTER TO REPLACE TOP 5 AND LOWER 5 PERCENT VALUES

def replace_percentiles(column):
    
    p5 = np.percentile(column, 5)
    p95 = np.percentile(column, 95)
   
    is_top = column > np.percentile(column, 95)
    is_bottom = column < np.percentile(column, 5)
    
    column.loc[is_top] = p95
    column.loc[is_bottom] = p5
    
    return column

replace_percentiles(data['fruitset'])
replace_percentiles(data['fruitmass'])
replace_percentiles(data['yield'])
replace_percentiles(data['seeds'])
replace_percentiles(test['fruitset'])
replace_percentiles(test['fruitmass'])
replace_percentiles(test['seeds'])
```

```python
# FREQUENCY ENCODING THE CATEGORICAL VARIABLES AS NO. OF CLASSES IS HIGH

def frequency_encode_columns(data, cols):
    for col in cols:
        freq = data[col].value_counts(normalize=True)
        data[col] = data[col].map(freq)
    return data

# Applying frequency encoding to the 'data' and 'test' dataframes
data = frequency_encode_columns(data,['clonesize','honeybee','bumbles','andrena','osmia','MaxOfUpperTRange','MinOfUpperTRange','AverageOfUpperTRange','MaxOfLowerTRange','MinOfLowerTRange','AverageOfLowerTRange','RainingDays','AverageRainingDays'])
test = frequency_encode_columns(test,['clonesize','honeybee','bumbles','andrena','osmia','MaxOfUpperTRange','MinOfUpperTRange','AverageOfUpperTRange','MaxOfLowerTRange','MinOfLowerTRange','AverageOfLowerTRange','RainingDays','AverageRainingDays'])
```

```python
# NORMALIZING CONTINOUS VARIABLES

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
continuous_cols = ['fruitset', 'fruitmass', 'seeds']

for col in continuous_cols:
    data[col] = scaler.fit_transform(data[[col]])
    test[col] = scaler.fit_transform(test[[col]])
```

```python
# CHECKING RESULT AFTER DATA NORMALIZATION

data.head()
```

```python
# CREATING BINS FOR CONTINOUS VARIABLES 

from sklearn.preprocessing import KBinsDiscretizer

discretizer = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')

continous_cols = ['fruitset', 'fruitmass', 'seeds']

for col in continous_cols:
    col_data = data[col].values.reshape(-1, 1)
    test_col_data= test[col].values.reshape(-1,1)
    
    col_discretized = discretizer.fit_transform(col_data)
    test_col_discretised = discretizer.fit_transform(test_col_data)
    
    col_discretized = pd.DataFrame(col_discretized, columns=[col])
    test_col_discretised = pd.DataFrame(test_col_discretised, columns=[col])
    
    data[col] = np.ravel(col_discretized)
    test[col] = np.ravel(test_col_discretised)
```
